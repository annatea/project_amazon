```python

all_files = glob.glob(os.path.join(path, "*.csv"))

all_df = []
for f in all_files:
    df = pd.read_csv(f, sep=',')
    df['file'] = f.split('/')[-1]
    all_df.append(df)
    
merged_df = pd.concat(all_df, ignore_index=True, sort=True)
```

```python
merged_df
```

```python
merged_df["Brand"] = merged_df['file'].map(lambda x: x.lstrip('amazon\amazon_').rstrip('.csv'))
```

```python
merged_df['Brand'] = merged_df['Brand'].replace('\amazon_', ' ')
merged_df['Brand']
```
